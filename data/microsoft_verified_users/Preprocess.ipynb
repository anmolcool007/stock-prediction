```python
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r"http\S+", "", x))
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r"www.+", "", x))
```

```python
df['tweet'] = df['tweet'].apply(lambda x: re.sub("#", "", x))
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r"@+", "", x))
ll = df['tweet'].tolist()
```
